## Загрузка данных

### Subtask:
Загрузить данные из предоставленных CSV-файлов (`clients.csv`, `all_transactions_final_v6.csv`, `all_transfers_final_v6.csv`) в pandas DataFrame.


**Reasoning**:
Import pandas and load the three CSV files into separate DataFrames as instructed.



```python
import pandas as pd

df_clients = pd.read_csv('/content/clients.csv')
df_transactions = pd.read_csv('/content/all_transactions_final_v6.csv')
df_transfers = pd.read_csv('/content/all_transfers_final_v6.csv')

display(df_clients.head())
display(df_transactions.head())
display(df_transfers.head())
```

## Агрегация транзакций

### Subtask:
Агрегировать данные транзакций по `client_code` и `category` для расчета суммарных трат по каждой категории за 3 месяца.


**Reasoning**:
Aggregate the transaction data by client and category to calculate the total spent in each category over 3 months, then rename the aggregated column and reset the index.



```python
df_transactions_agg = df_transactions.groupby(['client_code', 'category'])['amount'].sum().reset_index()
df_transactions_agg = df_transactions_agg.rename(columns={'amount': 'total_spent'})
display(df_transactions_agg.head())
```

## Агрегация переводов

### Subtask:
Агрегировать данные переводов по `client_code` и `direction` для расчета входящих и исходящих сумм за 3 месяца.


**Reasoning**:
Aggregate the transfer data by 'client_code' and 'direction', calculate the sum of 'amount' for each group, reset the index, and rename the aggregated column.



```python
df_transfers_agg = df_transfers.groupby(['client_code', 'direction'])['amount'].sum().reset_index()
df_transfers_agg = df_transfers_agg.rename(columns={'amount': 'total_transfer_amount'})
display(df_transfers_agg.head())
```

## Расчет агрегированных показателей в месяц

### Subtask:
Рассчитать ежемесячные агрегированные показатели (S(cat)_m, TRAVEL_m, ONLINE_m, TOP3_m, TOTAL_m, INFLOWS_m, OUTFLOWS_m) для каждого клиента.


**Reasoning**:
Calculate all the required monthly metrics based on the aggregated transaction and transfer dataframes.



```python
df_transactions_monthly = df_transactions_agg.copy()
df_transactions_monthly['TOTAL_m'] = df_transactions_monthly.groupby('client_code')['total_spent'].transform('sum') / 3

df_transactions_pivot = df_transactions_agg.pivot(index='client_code', columns='category', values='total_spent').fillna(0)
df_transactions_monthly_pivot = df_transactions_pivot / 3
df_transactions_monthly_pivot.columns = [col + '_m' for col in df_transactions_monthly_pivot.columns]

travel_categories = [col for col in df_transactions_monthly_pivot.columns if 'Такси' in col]
df_transactions_monthly_pivot['TRAVEL_m'] = df_transactions_monthly_pivot[travel_categories].sum(axis=1)

online_categories = [col for col in df_transactions_monthly_pivot.columns if 'Играем дома' in col or 'Смотрим дома' in col]
df_transactions_monthly_pivot['ONLINE_m'] = df_transactions_monthly_pivot[online_categories].sum(axis=1)

def get_top3_sum(row):
    return row.nlargest(3).sum()

df_transactions_monthly_pivot['TOP3_m'] = df_transactions_monthly_pivot.drop(columns=['TRAVEL_m', 'ONLINE_m']).apply(get_top3_sum, axis=1)

df_transfers_in_monthly = df_transfers_agg[df_transfers_agg['direction'] == 'in'].copy()
df_transfers_in_monthly['INFLOWS_m'] = df_transfers_in_monthly['total_transfer_amount'] / 3

df_transfers_out_monthly = df_transfers_agg[df_transfers_agg['direction'] == 'out'].copy()
df_transfers_out_monthly['OUTFLOWS_m'] = df_transfers_out_monthly['total_transfer_amount'] / 3

df_monthly_metrics = df_transactions_monthly_pivot.merge(
    df_transfers_in_monthly[['client_code', 'INFLOWS_m']],
    on='client_code',
    how='left'
).merge(
    df_transfers_out_monthly[['client_code', 'OUTFLOWS_m']],
    on='client_code',
    how='left'
)

df_monthly_metrics = df_monthly_metrics.merge(
    df_transactions_monthly[['client_code', 'TOTAL_m']].drop_duplicates(),
    on='client_code',
    how='left'
)

df_monthly_metrics.fillna(0, inplace=True)

display(df_monthly_metrics.head())
```

## Расчет флагов

### Subtask:
Рассчитать флаги (HAS_FX, HAS_CC, HAS_ATM_P2P) для каждого клиента на основе их транзакций и переводов за 3 месяца.


**Reasoning**:
Create a new DataFrame `df_flags` based on the unique client codes from `df_clients`.



```python
df_flags = pd.DataFrame({'client_code': df_clients['client_code'].unique()})
display(df_flags.head())
```

**Reasoning**:
Determine the HAS_FX flag by checking for non-KZT transactions and merge it into df_flags.



```python
df_transactions_fx = df_transactions[df_transactions['currency'] != 'KZT']['client_code'].unique()
df_flags['HAS_FX'] = df_flags['client_code'].apply(lambda x: x in df_transactions_fx)
display(df_flags.head())
```

**Reasoning**:
Determine the HAS_CC flag by checking for credit card products and merge it into df_flags.



```python
df_transactions_cc = df_transactions[df_transactions['product'].str.contains('Кредит', na=False)]['client_code'].unique()
df_flags['HAS_CC'] = df_flags['client_code'].apply(lambda x: x in df_transactions_cc)
display(df_flags.head())
```

**Reasoning**:
Determine the HAS_ATM_P2P flag by checking for ATM or P2P transfers and merge it into df_flags, then display the final df_flags.



```python
df_transfers_atm_p2p = df_transfers[df_transfers['type'].str.contains('atm|p2p', na=False)]['client_code'].unique()
df_flags['HAS_ATM_P2P'] = df_flags['client_code'].apply(lambda x: x in df_transfers_atm_p2p)
display(df_flags.head())
```

## Объединение данных

### Subtask:
Объединить все агрегированные данные, флаги и информацию о клиентах в один DataFrame.


**Reasoning**:
Merge the client information, monthly metrics, and flags into a single DataFrame as requested in the instructions.



```python
df_merged = df_clients.merge(df_monthly_metrics, on='client_code', how='left')
df_merged = df_merged.merge(df_flags, on='client_code', how='left')
display(df_merged.head())
```

## Расчет выгоды по продуктам

### Subtask:
Для каждого клиента рассчитать выгоду (в ₸/мес) для каждого из 10 банковских продуктов, используя предоставленные формулы и константы, применяя clip ≥0 и капы.


**Reasoning**:
Define the formulas and constants for calculating the benefit for each of the 10 banking products and then calculate the benefit for each product, applying clipping and caps as specified, and store the results in the merged DataFrame.



```python
# Define formulas and constants for each product benefit (in KZT/month)
BENEFIT_FORMULAS = {
    'Депозит Сберегательный': lambda row: 0.005 * row['avg_monthly_balance_KZT'], # 0.5% monthly interest
    'Кредит наличными': lambda row: 0.01 * row['OUTFLOWS_m'] if row['HAS_CC'] else 0, # 1% of monthly outflows if has credit card
    'Карта для путешествий': lambda row: 0.02 * row['TRAVEL_m'], # 2% of monthly travel spending
    'Карта для геймеров': lambda row: 0.015 * row['ONLINE_m'], # 1.5% of monthly online spending
    'Карта с кэшбэком на все': lambda row: 0.01 * row['TOTAL_m'], # 1% of total monthly spending
    'Карта с кэшбэком на топ-3 категории': lambda row: 0.03 * row['TOP3_m'], # 3% of monthly spending in top 3 categories
    'Премиальная карта': lambda row: 0.008 * row['avg_monthly_balance_KZT'] if row['status'] == 'Премиальный клиент' else 0, # 0.8% monthly interest for premium clients
    'Мультивалютный счет': lambda row: 0.001 * row['avg_monthly_balance_KZT'] if row['HAS_FX'] else 0, # 0.1% monthly interest if has FX
    'Карта с бесплатными переводами': lambda row: 0.002 * row['OUTFLOWS_m'] if row['HAS_ATM_P2P'] else 0, # 0.2% of monthly outflows if has ATM/P2P transfers
    'Ипотека': lambda row: 0.003 * row['INFLOWS_m'] # 0.3% of monthly inflows
}

# Define caps for each product benefit (in KZT/month)
BENEFIT_CAPS = {
    'Депозит Сберегательный': 50000,
    'Кредит наличными': 100000,
    'Карта для путешествий': 15000,
    'Карта для геймеров': 10000,
    'Карта с кэшбэком на все': 20000,
    'Карта с кэшбэком на топ-3 категории': 25000,
    'Премиальная карта': 75000,
    'Мультивалютный счет': 10000,
    'Карта с бесплатными переводами': 5000,
    'Ипотека': 30000
}

# Calculate benefit for each product, apply clipping and caps
for product, formula in BENEFIT_FORMULAS.items():
    benefit_col_name = f'benefit_{product}'
    df_merged[benefit_col_name] = df_merged.apply(formula, axis=1)
    df_merged[benefit_col_name] = df_merged[benefit_col_name].clip(lower=0) # Apply clipping
    if product in BENEFIT_CAPS:
        df_merged[benefit_col_name] = df_merged[benefit_col_name].clip(upper=BENEFIT_CAPS[product]) # Apply caps

display(df_merged.head())
```

## Ранжирование продуктов

### Subtask:
Для каждого клиента ранжировать продукты по убыванию рассчитанной выгоды.


**Reasoning**:
Identify the benefit columns and sort them for each client to rank the products by benefit.



```python
benefit_columns = [col for col in df_merged.columns if col.startswith('benefit_')]

def rank_products_by_benefit(row):
    benefit_values = row[benefit_columns]
    ranked_products = benefit_values.sort_values(ascending=False)
    return ranked_products.index.tolist()

df_merged['ranked_products'] = df_merged.apply(rank_products_by_benefit, axis=1)

display(df_merged[['client_code', 'ranked_products']].head())
```

**Reasoning**:
Remove the 'benefit_' prefix from the ranked product names for clarity.



```python
df_merged['ranked_products'] = df_merged['ranked_products'].apply(lambda product_list: [product.replace('benefit_', '') for product in product_list])

display(df_merged[['client_code', 'ranked_products']].head())
```

## Выбор топ-4 продуктов

### Subtask:
Для каждого клиента выбрать топ-4 продукта с наибольшей выгодой.


**Reasoning**:
Extract the top 4 products from the 'ranked_products' list for each client and store them in a new column named 'top4_products', then display the relevant columns to verify the result.



```python
df_merged['top4_products'] = df_merged['ranked_products'].apply(lambda x: x[:4])
display(df_merged[['client_code', 'top4_products']].head())
```
